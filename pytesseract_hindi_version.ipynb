```python
# Install the googletrans library
!pip install googletrans==4.0.0-rc1

# Import the Translator class from googletrans
from googletrans import Translator

# Create a Translator object
translator = Translator()

# Function to translate text from Hindi to English
def translate_hindi_to_english(text):
    translated_text = translator.translate(text, src='hi', dest='en').text
    return translated_text

# Function to translate text from English to Hindi
def translate_english_to_hindi(text):
    translated_text = translator.translate(text, src='en', dest='hi').text
    return translated_text

# Test the translation functions
hindi_text = "नमस्ते, क्या हाल है?"
english_translation = translate_hindi_to_english(hindi_text)
hindi_translation = translate_english_to_hindi(english_translation)

print("Hindi Text:", hindi_text)
print("English Translation:", english_translation)
print("Hindi Translation:", hindi_translation)
```
